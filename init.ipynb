```python
import pandas as pd
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
```

```python
x_res = []
x_y = []
x_mean, y_mean = x.mean(), y.mean()
for i in range(x.shape[0]):
    x_res.append( (x[i]-x_mean)**2 )
    x_y.append((x[i]-x_mean)*(y[i]-y_mean))
x_res, x_y
```

```python
b1 = sum(x_y)/sum(x_res)
b1
```

```python
b_0 = y_mean - (b1 * x_mean)
b_0
```

```python
import matplotlib.pyplot as plt

x_ = np.linspace(-5,5,100)
y_ = b_0 + b1 * x_
plt.scatter(x, y, color='g')
plt.plot(x_,y_, color='black')
plt.show()
```
